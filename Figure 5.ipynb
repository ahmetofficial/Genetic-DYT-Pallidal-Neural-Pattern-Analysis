```python
import pandas as pd
import numpy as np
import math
from os.path import exists
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from decimal import Decimal
from matplotlib.colors import BoundaryNorm
from matplotlib.patches import Rectangle

from scipy import stats
import statsmodels.stats.multitest
from decimal import Decimal
import collections

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

import pickle 
from matplotlib import rcParams

import warnings
warnings.filterwarnings('ignore')

import utils_hyperparameter_optimizer, utils_machine_learning
```

# 1. Data Loading

```python
dystonia_genetics = pd.read_csv('dataset/genetic_dystonia_pallidal_neuron_activity.csv')
dystonia_genetics = dystonia_genetics[dystonia_genetics.patient!="patient1"] # removing benign-SGCE case

# fill np.nan values with 0
dystonia_genetics = dystonia_genetics.fillna(0)

# these two features can have infinite values due to 0 division problem. while their measurement
dystonia_genetics['pause_index'].replace(np.inf,dystonia_genetics.loc[dystonia_genetics['pause_index'] != np.inf, 'pause_index'].max(),inplace=True)
dystonia_genetics['pause_ratio'].replace(np.inf,dystonia_genetics.loc[dystonia_genetics['pause_ratio'] != np.inf, 'pause_ratio'].max(),inplace=True)

gene_map = {'AOPEP':1, 'GNAL':2, 'KMT2B':3, 
            'PANK2':4, 'PLA2G6':5, 'SGCE' :6, 
            'THAP1':7, 'TOR1A' :8, 'VPS16':9}
```

# 2. OVO Classification

In this part of the analysis, we try to classify the neurons belonging to a specific genetic factor with excluding neurons that belongs to idiopathic dystonia patients

```python
mann_wu_results  = pd.read_csv("test_results/mann_whitney_u_results.csv")
fisher_e_results = pd.read_csv("test_results/fisher_e_results.csv")

features         = list(mann_wu_results.biomarker.unique())
features.extend(fisher_e_results.biomarker.unique())
```

## 2.1. Preprocessing

```python
X        = dystonia_genetics[features]
Y        = dystonia_genetics.gene.replace(gene_map).astype(np.int32)
X        = X.fillna(0)

scaler   = StandardScaler()

X_scaled = scaler.fit_transform(X)
X        = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)
```

```python
features_OVO      = {}
unique_gene_pairs = pd.DataFrame(mann_wu_results[['gene1', 'gene2']]).apply(lambda x: tuple(sorted(x)), axis=1).unique()

for gene_pair in unique_gene_pairs:
    gene1               = gene_pair[0]
    gene2               = gene_pair[1]

    features_for_pairs  = list(mann_wu_results[(mann_wu_results.gene1==gene1) & (mann_wu_results.gene2==gene2) & (mann_wu_results.pvalue<=0.05)].biomarker.unique())
    features_for_pairs.extend(fisher_e_results[(fisher_e_results.gene1==gene1) & (fisher_e_results.gene2==gene2) & (fisher_e_results.pvalue<=0.05)].biomarker.unique())

    if(len(features_for_pairs)==0): #if gene pair does not have significantly differed features, consider all features for classification
        features_OVO[gene_pair] = features
    else:
        features_OVO[gene_pair] = features_for_pairs
```

## 2.2. Training

```python
performance_matrix = pd.DataFrame(columns=["model", "analysis", "balanced_accuracy", "weighted_auc", "weighted_f1"])
CI_matrix          = pd.DataFrame(columns=["model", "analysis", "balanced_accuracy_UB", "weighted_auc_UB", "weighted_f1_UB"])

for gene_pair in unique_gene_pairs:
    gene1 = gene_pair[0]
    gene2 = gene_pair[1]

    print("------------------------------------------------------------------------")
    print("Model training for: " + gene1 + " - " + gene2 + " pair started...")
    print("------------------------------------------------------------------------")

    # prepare the X, Y data structures for each gene pair
    X_gene_pair = pd.DataFrame(X[(Y==gene_map[gene1]) | (Y==gene_map[gene2])])[features_OVO[gene_pair]]
    Y_gene_pair = Y.copy()[(Y==gene_map[gene1]) | (Y==gene_map[gene2])]
    pair        = gene1 + "_" + gene2

    #########################################################################################
    #########################################################################################
    #########################################################################################
    
    print("> Random Forest Classifier Model")

    params_rf                     = utils_machine_learning.get_optimized_parameters(X_gene_pair, Y_gene_pair, classification_mode="ovo", model="rf", gene_pair=pair, params_extra=None)
    model_rf,  performance_matrix = utils_machine_learning.get_model_output(X_gene_pair, Y_gene_pair, params_rf, "rf", pair, classes=[gene1,gene2], performance=performance_matrix)
    pickle.dump(model_rf, open("trained_models/ovo/rf_"+pair, 'wb'))
    
    print("---> model trained...")
    print("---> CI estimation started...")
    
    acc_threshold, auc_threshold, f1_threshold = utils_machine_learning.get_model_CI(X_gene_pair, Y_gene_pair, params_rf, "rf", iteration=10, 
                                                                                     model_all=np.NaN, model_pair=np.NaN)

    CI_row = {"model":"rf", "analysis": pair, "balanced_accuracy_UB":acc_threshold, "weighted_auc_UB":auc_threshold, "weighted_f1_UB":f1_threshold}
    CI_matrix.loc[len(CI_matrix)] = CI_row
    #########################################################################################
    #########################################################################################
    #########################################################################################

    print("> Gradient Boosting Classifier Model")
    params_gbc                    = utils_machine_learning.get_optimized_parameters(X_gene_pair, Y_gene_pair, classification_mode="ovo", model="gbc", gene_pair=pair, params_extra=params_rf)
    model_gbc, performance_matrix = utils_machine_learning.get_model_output(X_gene_pair, Y_gene_pair, params_gbc, "gbc", pair, classes=[gene1,gene2], performance=performance_matrix)
    pickle.dump(model_gbc, open("trained_models/ovo/gbc_"+pair, 'wb'))
    
    print("---> model trained...")
    print("---> CI estimation started...")
    
    acc_threshold, auc_threshold, f1_threshold = utils_machine_learning.get_model_CI(X_gene_pair, Y_gene_pair, params_gbc, "gbc", iteration=10, 
                                                                                     model_all=np.NaN, model_pair=np.NaN)

    CI_row = {"model":"gbc", "analysis": pair, "balanced_accuracy_UB":acc_threshold, "weighted_auc_UB":auc_threshold, "weighted_f1_UB":f1_threshold}
    CI_matrix.loc[len(CI_matrix)] = CI_row

    #########################################################################################
    #########################################################################################
    #########################################################################################
    
    print("> Voting Classifier Model")
    models_for_vc        = dict()
    models_for_vc["rf"]  = model_rf
    models_for_vc["gbc"] = model_gbc
            
    model_vc, performance_matrix = utils_machine_learning.get_model_output(X_gene_pair, Y_gene_pair, np.NaN, "vc", pair, classes=[gene1,gene2], performance=performance_matrix, model_all=models_for_vc, model_pair=["rf","gbc"])
    pickle.dump(model_vc, open("trained_models/ovo/vc_"+pair, 'wb'))
    
    print("---> model trained...")
    print("---> CI estimation started...")

    acc_threshold, auc_threshold, f1_threshold = utils_machine_learning.get_model_CI(X_gene_pair, Y_gene_pair, np.NaN, "vc", iteration=10, model_all=models_for_vc, model_pair=["rf","gbc"])

    CI_row = {"model":"vc", "analysis": pair, "balanced_accuracy_UB":acc_threshold, "weighted_auc_UB":auc_threshold, "weighted_f1_UB":f1_threshold}
    CI_matrix.loc[len(CI_matrix)] = CI_row
```

```python
performance_matrix.to_csv("model_results/ovo_performance_matrix.csv", index=False)
CI_matrix.to_csv("model_results/CI_matrix.csv", index=False)
```

# 3. OVR Classification

```python
stats_OVR           = dict()
stats_OVR["AOPEP"]  = pd.read_csv("test_results/one-vs-rest/AOPEP_vs_Rest.csv")
stats_OVR["GNAL"]   = pd.read_csv("test_results/one-vs-rest/GNAL_vs_Rest.csv")
stats_OVR["KMT2B"]  = pd.read_csv("test_results/one-vs-rest/KMT2B_vs_Rest.csv")
stats_OVR["PANK2"]  = pd.read_csv("test_results/one-vs-rest/PANK2_vs_Rest.csv")
stats_OVR["PLA2G6"] = pd.read_csv("test_results/one-vs-rest/PLA2G6_vs_Rest.csv")
stats_OVR["SGCE"]   = pd.read_csv("test_results/one-vs-rest/SGCE_vs_Rest.csv")
stats_OVR["THAP1"]  = pd.read_csv("test_results/one-vs-rest/THAP1_vs_Rest.csv")
stats_OVR["TOR1A"]  = pd.read_csv("test_results/one-vs-rest/TOR1A_vs_Rest.csv")
stats_OVR["VPS16"]  = pd.read_csv("test_results/one-vs-rest/VPS16_vs_Rest.csv")

features            = list(stats_OVR["AOPEP"].biomarker.unique())
```

## 3.1. Preprocessing

```python
X        = dystonia_genetics[features]
Y        = dystonia_genetics.gene.replace(gene_map).astype(np.int32)
X        = X.fillna(0)

scaler   = StandardScaler()

X_scaled = scaler.fit_transform(X)
X        = pd.DataFrame(X_scaled, index=X.index, columns=X.columns)
```

```python
features_OVR = {}

for gene in stats_OVR.keys():

    features_gene_OVR = list(stats_OVR[gene][stats_OVR[gene].pvalue<=0.05].biomarker.unique())

    if(len(features_gene_OVR)==0): #if gene pair does not have significantly differed features, consider all features for classification
        features_OVR[gene] = features
    else:
        features_OVR[gene] = features_gene_OVR
```

## 3.2 Training

### 3.2.1 Measure Performances

```python
performance_matrix = pd.DataFrame(columns=["model", "analysis", "balanced_accuracy", "weighted_auc", "weighted_f1"])
cm_OVR             = dict()

for gene in stats_OVR.keys():

    print("------------------------------------------------------------------------")
    print("Model training for: " + gene + " vs REST started...")
    print("------------------------------------------------------------------------")
    
    X_gene_vs = pd.DataFrame(X[features_OVR[gene]])
    Y_gene_vs = Y.copy().map(lambda x: 1 if x == gene_map[gene] else 0)
    pair      = gene+"_vs_REST"

    #########################################################################################
    #########################################################################################
    #########################################################################################
    
    print("> Random Forest Classifier Model")

    params_rf                     = utils_machine_learning.get_optimized_parameters(X_gene_vs, Y_gene_vs, classification_mode="ovr", model="rf", gene_pair=pair, params_extra=None)
    model_rf,  performance_matrix = utils_machine_learning.get_model_output(X_gene_vs, Y_gene_vs, params_rf, "rf", pair, classes=[gene,"REST"], performance=performance_matrix)
    pickle.dump(model_rf, open("trained_models/ovr/rf_"+pair, 'wb'))
    
    print("---> model trained...")

    #########################################################################################
    #########################################################################################
    #########################################################################################

    print("> Gradient Boosting Classifier Model")
    params_gbc                    = utils_machine_learning.get_optimized_parameters(X_gene_vs, Y_gene_vs, classification_mode="ovr", model="gbc", gene_pair=pair, params_extra=params_rf)
    model_gbc, performance_matrix = utils_machine_learning.get_model_output(X_gene_vs, Y_gene_vs, params_gbc, "gbc", pair, classes=[gene,"REST"], performance=performance_matrix)
    pickle.dump(model_gbc, open("trained_models/ovr/gbc_"+pair, 'wb'))
    
    print("---> model trained...")

    #########################################################################################
    #########################################################################################
    #########################################################################################
    
    print("> Voting Classifier Model")
    models_for_vc        = dict()
    models_for_vc["rf"]  = model_rf
    models_for_vc["gbc"] = model_gbc
            
    model_vc, performance_matrix = utils_machine_learning.get_model_output(X_gene_vs, Y_gene_vs, np.NaN, "vc", pair, classes=[gene,"REST"], performance=performance_matrix, model_all=models_for_vc, model_pair=["rf","gbc"])
    pickle.dump(model_vc, open("trained_models/ovr/vc_"+pair, 'wb'))
    
    print("---> model trained...")

    cm_OVR[gene]                 = utils_machine_learning.get_model_confusion_matrix(X_gene_vs, Y_gene_vs, np.NaN, "vc", pair, classes=[gene,"REST"], performance=performance_matrix, model_all=models_for_vc, model_pair=["rf","gbc"])


performance_matrix.to_csv("model_results/ovr_performance_matrix.csv", index=False)
pickle.dump(cm_OVR, open("model_results/ovr_CM", 'wb'))
```
